```python
def movie_genre(detail_link): # we are defining this function to get the genre information of all the 500 movies from metacritics
    detail_page = requests.get(detail_link, headers = headers) 
    detail_soup = BeautifulSoup(detail_page.content, "html.parser")
    genres=detail_soup.select('tr.genres td.data span') # I observed that this code is perfect to fetch the genre information from the webpages using the tags mentioned by using select() fuction
    final_genre = [] # creating an empty list to store the values from the below code
    for genre in genres: # using for loop and looping through the webpage by using the variable that stored the tags to find genre informations from webpages for all 500 movies
        info = genre.get_text() # this variable stores the text from the tags 
        final_genre.append(info.strip()) # this code appends the genre information to the list for each movies using for loop
    return final_genre # this will help us to return the result of the function when its called in the main program
```

```python
#let's then try the get cast function

def get_cast(detail_link): # we are defining this function to get the Principal casts information of all the 500 movies from metacritics
    detail_page = requests.get(detail_link, headers = headers)
    detail_soup = BeautifulSoup(detail_page.content, "html.parser")
    try:
        cast_table = detail_soup.find('table', class_ = 'credits', summary = movie_name + "'s Cast Credits") # the class and summary, and table tag mentioned is unique and used for finding the casts info that has "'s'" along with the movie name in webpage
        cast_line = cast_table.find_all('td', class_ = 'person')# this code locates the td tag and class mentioned, under the above mentioned position in the webpage
    except:
        try:
            cast_table = detail_soup.find('table', class_ = 'credits', summary = movie_name + "' Cast Credits") # the class and summary, and table tag mentioned is unique and used for finding the casts info that doesn't has "'s'" along with the movie name in webpage
            cast_line = cast_table.find_all('td', class_ = 'person')# this code locates the td tag and class mentioned, under the above mentioned position in the webpage
        except:
            cast_line = None # if some movie doesn't containes cast information, this block will be executed 
            
    casts = []  # creating an empty list to store the values from the below code  
    if cast_line == None: # if any movie doesn't containes casts information, empty list will be added for that particular movie
        casts =[]
    else:
        for line in cast_line: # if the casts value is found for loop starts
            hyperlink = line.find('a')# we are finding the hyperlink from the variable that containes located position for the casts from the webpages
            cast = hyperlink.get_text() # this code is used to get the text from the hyperlink and stores it in a variable
            casts.append(cast.strip()) # this code appends the result from the above code to a list by striping off the spaces if any
    return casts # this will help us to return the result of the function when its called in the main program
```

```python
# now we are getting started with the main function
all_movie = [] # creating an empty list to store the final result of the code
import time # this is used to execute any time operations used in the code
n = 0 # counting to '0' to assign the number for movie name as the program starts
for movie in container: # using for loop to loop through the variable container
    n = n+1 # now n adds to '1'
    print(n) # this will print the value of 'n' for each loop
    movie_name = get_movie_name() # calling this function to get movie names and storing it in variable
    print(movie_name) # printing the name of the movie for each loop
    movie_link_tag = movie.find('a', class_ = 'title') #using this tag, we are getting the link of each movie name
    movie_link = movie_link_tag.get('href', None) # this is used to fetch the link from the above code 
    movie_link = 'https://www.metacritic.com'+ movie_link # this adds the content 'https://www.metacritic.com' for each movie link and stores it in a variable in order to use to fetch data from this web page in further programs  
    detail_page = requests.get(movie_link, headers = headers) 
    detail_soup = BeautifulSoup(detail_page.content, "html.parser")
    div = detail_soup('div', class_= 'see_all pad_btm1') # this is the tag for locating 'See All Details and Credits' in the movie link page
    for line in div:
        hyperlink = line.find('a') # this is used to locate the link tag from the above code
        detail_link = hyperlink.get('href', None) # this code is used to get the link and stores it in a variable
        detail_link = 'https://www.metacritic.com'+ detail_link # this code adds the content 'https://www.metacritic.com' and stores in a variable
        #now we have the detail link for each movie, again, we are only dealing with one movie here
        print(detail_link) # this prints the links of the detail page
        genre = movie_genre(detail_link) # calling this function to get genre information for each and storing it in variable
        print(genre) 
        
        p_cast = get_p_cast(detail_link) # calling this function to get principal cast information for each and storing it in variable
        print(p_cast)
        
        cast = get_cast(detail_link) # calling this function to get other cast information for each and storing it in variable
        print(cast)
        time.sleep(1) 
        combined_casts = p_cast + cast # concatenating the casts and principal cast 
        print(combined_casts) # printing the overall casts information for each movie
    cast_genre = dict() # creating a dictionary to store cast information as keys, and genre information as values
    cast_genre[tuple(combined_casts)] = genre 
    movie_cast = dict() # creating a dictionary to store movie information as keys, and cast& genre informations as values
    movie_cast[movie_name] = cast_genre
    print(movie_cast)
    all_movie.append(movie_cast) # appending the final dictionary named movie_cast to the very first list created in the main function
print(all_movie) # finally printing the {movie_name:{cast information: genre information}} 
     
```
